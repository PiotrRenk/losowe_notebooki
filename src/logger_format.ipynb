```python
import logging
```

### Basic setup:

```python
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s  %(levelname)-8s  %(name)s — %(message)s",
)
logging.info("Logging is configured.")
logger = logging.getLogger(__name__)
```

notes: 
- % is for format of the string
- %s is a placeholder for string
- %d is a placeholder for int

```python
logger.info("This is an info message.")
```

```python
logger.warning("This is a warning message.")
```

```python
username = "%s"
logger.info("Difference between %s", username)
```

```python
username = "f-string{}"
logger.info(f"and formatting using f-strings {username}")
```

```python
logger.info("Is that % waits for formating until logger is emmited, while the f-string is evaluated immediately, and that the value is passed")
```

```python
logger = logging.getLogger(__name__)

#one logger to log and displey in the console
console_handler = logging.StreamHandler()
#set level will display INFO and above (Warning, Error, Critical) in the console
console_handler.setLevel(logging.INFO)
#sets the format of the return messages
console_handler.setFormatter("%(levelname)s, %(message)s")

#and another logger for saving to    the file:
file_handler = logging.FileHandler("../utils/logging_file.log")
#this will also have Debug ontop
file_handler.setLevel(logging.DEBUG)
console_handler.setFormatter('%(asctime)s::%(levelname)s::'
                            '%(name)s::%(funcName)s::'
                            '%(lineno)d::(message)s')
logger.addHandler(console_handler)
logger.addHandler(file_handler)
```
